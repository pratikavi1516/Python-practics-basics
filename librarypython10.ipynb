# Pie chart

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1600,600,300,410,250]
exp_labels = ["House_rent", "Food", "Phone/Internet bill", "car", "other utilities"]
```

```python
plt.pie(exp_vals,labels= exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 2)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow=True, explode=[0,0.5,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow=True,startangle=90)
plt.show()
```

# how to save matplotlib file

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow=True,explode=[0,0.5,0,0,0],startangle=90)
plt.show()
plt.savefig("piechart.png")
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow=True,explode=[0,0.5,0,0,0],startangle=90)
plt.show()
plt.savefig("spiechart.png", bbox_inches= "tight", pad_inches= 2)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels= exp_labels, radius= 1.5, autopct= '%0.1f%%', shadow=True,explode=[0,0.5,0,0,0],startangle=90)
plt.savefig("dpiechart.jpeg", bbox_inches= "tight", pad_inches= 2)
```
