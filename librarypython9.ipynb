# bar diagram

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company =['GOOGLE','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
profit= [40,2,34,12]
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
plt.bar(xpos,profit)
```

```python
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue, width=0.4)
plt.bar(xpos+0.2,profit, width=0.4)
```

```python
plt.yticks(xpos,company)
plt.barh(xpos-0.2,revenue)
plt.barh(xpos+0.2,profit)
```

# Histogram

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color= 'gold')
```

```python
blood_sugar_man = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,130,84,69,89,79,120,112,100]
plt.hist([blood_sugar_man ,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['gold','green'] )
```

```python
plt.hist([blood_sugar_man ,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['gold','green'] , orientation= 'horizontal')
```
