# 2.5 Pandas Handle Missing data :isnull & not null,fillna,interpolate,dropna

#1) isnull  gives value True if cell is null otherwise False and not null gives converse
2) fillna to fill missing values using diff ways
3) interpolate to make a guess on missing values usinginterpolate
4)dropna to drop rows with missing values

```python
df=pd.read_excel (r'TEMP.xls')
```

```python
df.head()
```

```python
df.isnull().tail()
```

```python
df.notnull().tail()
```

```python
df.isnull().sum()   
```

```python
df.set_index('DAY', inplace=True)
df.head()
```

```python
new_df = df.fillna(0)  ##fill NaN values by 0
new_df.head()
```

```python
## new_df = df.fillna(0, inplace = True)   ## inplace method changes permanantly
```

```python
new_df = df['event'].fillna('RAIN')
new_df
```

```python
new_df = df.fillna(method= "ffill")
new_df.head()
```

```python
new_df = df.fillna(method= "bfill")
new_df.head()
```

```python
new_df = df.fillna(method= "pad")
new_df.head()
```

```python
new_df = df.fillna(method= "bfill", axis="columns")  #drop columns when atleast one element is missing
new_df.head()
```

```python
new_df = df.fillna(method= "ffill", limit=1)
new_df.head()
```

```python
new_df = df.interpolate()
new_df.head()
```

```python
new_df = df.interpolate(method ="time")
new_df.head()
```

```python
new_df = df.dropna()
new_df.head()
```

```python
##delete/ remove row if atleast one NaN is there

new_df = df.dropna(how="all")   ## how="all" if values are NaN drop that label
new_df.head()
```

```python
# new_df = df.dropna(thresh=1)
# new_df
```

```python
new_df = df.dropna(thresh=2)  #need 2 valid values
new_df.head()
```

```python
df.head()
```

```python
df.dropna(axis = 'columns')
```

```python
df.dropna(subset = ['event'])
```

```python
df.dropna(subset = ['event', 'windspeed'])
```

```python
new_df = df.replace([24.0,4.0, 'SNOW'], 'valid')
new_df.head()
```

```python
new_df = df.replace({
   'temperature' : 24.0,
    'windspeed' : 4.0,
    'event' : "SNOW"
}
, 'valid')
new_df.head()
```

```python
new_df = df.replace(
    {
   24.0 : 24.052145,
    7.0 : 7.58460,
    'RAIN' : "SNOW"
})
new_df.head()
```

```python
df6 =pd.read_excel (r'TEMP1.xls',encoding='cp1252')

df6.head()
```

```python
new_df = df6.replace('[A-Za-z]','',regex =True)

new_df.head()
```

```python
new_df = df6.replace({'temperature' : '[A-Za-z]', 'windspeed' : '[A-Za-z]'},'',regex =True)

new_df.head()
```
