# Pivot table

```python
df
```

```python
new_df = df.pivot(index="DAY", columns="city")
new_df
```

```python
dfa=pd.read_excel (r'TEMPERATURE.xls','Sheet3')

dfa
```

```python
new_df = dfa.pivot_table(index="DAY", columns="city")
new_df
```

```python
new_df = dfa.pivot_table(index="DAY", columns="city", aggfunc="sum")
new_df
```

```python
new_df = dfa.pivot_table(index="DAY", columns="city", aggfunc="count")
new_df
```

```python
new_df = dfa.pivot_table(index="DAY", columns="city", aggfunc="mean")
new_df
```

```python
new_df = dfa.pivot_table(index="DAY", columns="city", margins=True)
new_df
```

```python
df=pd.read_excel (r'TEMPERATURE.xls','Sheet2')

df
```

```python
new_df = df.pivot_table(index= pd.Grouper(freq= 'W', key = 'DAY'))        ## W stands for Week
new_df
```

```python
new_df = df.pivot_table(index= pd.Grouper(freq= 'M', key = 'DAY'))        ## M stands for Month
new_df
```
